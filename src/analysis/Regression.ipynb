```python
from pyspark.sql.functions import col, count, desc, when, isnan

# =========================================================
# PHẦN 1: KHÁM PHÁ DỮ LIỆU BẢNG CUSTOMERS
# =========================================================
print("================ KHÁM PHÁ BẢNG CUSTOMERS ================")
print("--- Cấu trúc dữ liệu ---")
customers.printSchema()
customers.show(5, truncate=False)

# Tính tổng số đơn và số khách hàng thực tế
total_orders = customers.count()
unique_customers = customers.select("customer_unique_id").distinct().count()

print(f"Tổng số lượt mua (customer_id): {total_orders}")
print(f"Tổng số khách hàng thực tế (customer_unique_id): {unique_customers}")
print(f"Số lượng khách hàng mua lại (Repeat buyers): {total_orders - unique_customers}")
print("-" * 40)

# Top 5 Bang có lượng khách hàng lớn nhất
print("--- Top 5 Bang đông khách hàng nhất ---")
customers.groupBy("customer_state") \
  .agg(count("customer_id").alias("total_customers")) \
  .orderBy(desc("total_customers")) \
  .show(5)

# Top 5 Thành phố có lượng khách hàng lớn nhất
print("--- Top 5 Thành phố đông khách hàng nhất ---")
customers.groupBy("customer_city") \
  .agg(count("customer_id").alias("total_customers")) \
  .orderBy(desc("total_customers")) \
  .show(5)

# Tìm và đếm những khách hàng mua hàng nhiều hơn 1 lần
print("--- Các khách hàng mua lại nhiều nhất ---")
repeat_buyers = customers.groupBy("customer_unique_id") \
  .agg(count("customer_id").alias("order_count")) \
  .filter(col("order_count") > 1) \
  .orderBy(desc("order_count"))
repeat_buyers.show(5, truncate=False)


# =========================================================
# PHẦN 2: LÀM SẠCH DỮ LIỆU BẢNG CUSTOMERS
# =========================================================
print("\n================ LÀM SẠCH BẢNG CUSTOMERS ================")

# 1. Kiểm tra và Xử lý Trùng lặp
print("--- 1. Xử lý Dữ liệu Trùng lặp ---")
# Mã định danh cho mỗi lần mua (customer_id) phải là duy nhất
customers_no_dups = customers.dropDuplicates(["customer_id"])

new_count = customers_no_dups.count()
print(f"Số dòng ban đầu: {total_orders}")
print(f"Số dòng sau khi xóa trùng lặp theo customer_id: {new_count}")
print(f"Đã phát hiện và xóa: {total_orders - new_count} dòng trùng lặp.\n")

# 2. Kiểm tra và Xử lý Null
print("--- 2. Kiểm tra Dữ liệu Null ---")
customers_no_dups.select([
    count(when(col(c).isNull() | isnan(c), c)).alias(c) for c in customers_no_dups.columns
]).show()

# Vì dữ liệu định vị (thành phố, bang) rất quan trọng trong bảng này,
# chúng ta sẽ điền "Unknown" cho các giá trị Null (nếu có) để không làm mất dòng dữ liệu.
customers_cleaned = customers_no_dups.fillna({
    "customer_zip_code_prefix": "00000",
    "customer_city": "Unknown",
    "customer_state": "Unknown"
})

final_count = customers_cleaned.count()
print(f"--- Hoàn tất! Số dòng dữ liệu customers_cleaned cuối cùng: {final_count} ---")
```

```python
from pyspark.sql.functions import col, count, desc, when, isnan

# =========================================================
# PHẦN 1: KHÁM PHÁ DỮ LIỆU BẢNG GEOLOCATION
# =========================================================
print("================ KHÁM PHÁ BẢNG GEOLOCATION ================")
print("--- Cấu trúc dữ liệu ---")
geolocation.printSchema()
geolocation.show(5, truncate=False)

# Tính tổng số dòng và số lượng mã bưu điện (zip code) duy nhất
total_geo_rows = geolocation.count()
unique_zip_codes = geolocation.select("geolocation_zip_code_prefix").distinct().count()

print(f"Tổng số dòng dữ liệu tọa độ: {total_geo_rows}")
print(f"Tổng số mã bưu điện (zip code) thực tế: {unique_zip_codes}")
print(f"Số lượng tọa độ bị dư thừa trên cùng 1 mã bưu điện: {total_geo_rows - unique_zip_codes}")
print("-" * 40)

# Xem 5 mã bưu điện nào có nhiều điểm tọa độ nhất
print("--- Top 5 mã bưu điện có nhiều điểm tọa độ nhất ---")
geolocation.groupBy("geolocation_zip_code_prefix") \
  .agg(count("*").alias("number_of_coordinates")) \
  .orderBy(desc("number_of_coordinates")) \
  .show(5)


# =========================================================
# PHẦN 2: LÀM SẠCH DỮ LIỆU BẢNG GEOLOCATION
# =========================================================
print("\n================ LÀM SẠCH BẢNG GEOLOCATION ================")

# 1. Kiểm tra và Xử lý Trùng lặp
print("--- 1. Xử lý Dữ liệu Trùng lặp ---")
# Như đã phân tích, để tạo thành một bảng tra cứu (lookup table) chuẩn xác 1-1,
# mỗi zip code chỉ nên có 1 dòng dữ liệu. Ta sẽ xóa trùng lặp dựa trên zip_code_prefix.
# (Hàm này sẽ giữ lại dòng đầu tiên nó gặp cho mỗi zip code và xóa các dòng sau)
geolocation_no_dups = geolocation.dropDuplicates(["geolocation_zip_code_prefix"])

new_geo_count = geolocation_no_dups.count()
print(f"Số dòng ban đầu: {total_geo_rows}")
print(f"Số dòng sau khi chuẩn hóa 1 zip code - 1 tọa độ: {new_geo_count}")
print(f"Đã dọn dẹp: {total_geo_rows - new_geo_count} dòng dữ liệu tọa độ thừa.\n")

# 2. Kiểm tra và Xử lý Null
print("--- 2. Kiểm tra Dữ liệu Null ---")
geolocation_no_dups.select([
    count(when(col(c).isNull() | isnan(c), c)).alias(c) for c in geolocation_no_dups.columns
]).show()

# Xử lý Null: Điền giá trị mặc định nếu có thành phố/bang bị thiếu
geolocation_cleaned = geolocation_no_dups.fillna({
    "geolocation_city": "Unknown",
    "geolocation_state": "Unknown"
})

final_geo_count = geolocation_cleaned.count()
print(f"--- Hoàn tất! Số dòng dữ liệu geolocation_cleaned cuối cùng: {final_geo_count} ---")
```

```python
from pyspark.sql.functions import col, count, desc, when, isnan, sum, max

# =========================================================
# PHẦN 1: KHÁM PHÁ DỮ LIỆU BẢNG ORDER ITEMS
# =========================================================
print("================ KHÁM PHÁ BẢNG ORDER ITEMS ================")
print("--- Cấu trúc dữ liệu ---")
order_items.printSchema()
order_items.show(5, truncate=False)

# Tính tổng số món hàng đã bán và số lượng đơn hàng thực tế
total_items_sold = order_items.count()
unique_orders = order_items.select("order_id").distinct().count()

print(f"Tổng số món hàng đã được bán ra: {total_items_sold}")
print(f"Tổng số lượng đơn hàng thực tế: {unique_orders}")
print(f"Trung bình mỗi đơn khách mua: {total_items_sold / unique_orders:.2f} món hàng")
print("-" * 40)

# Tính tổng doanh thu nền tảng (Tổng của cột price)
total_revenue = order_items.select(sum("price")).collect()[0][0]
print(f"--- TỔNG DOANH THU SẢN PHẨM: {total_revenue:.2f} BRL ---")

# Xem đơn hàng nào mua nhiều món hàng nhất (dựa vào số thứ tự max)
print("--- Top 5 đơn hàng mua nhiều sản phẩm nhất ---")
order_items.groupBy("order_id") \
  .agg(max("order_item_id").alias("total_items_in_order")) \
  .orderBy(desc("total_items_in_order")) \
  .show(5, truncate=False)

# Xem Top 5 sản phẩm bán chạy nhất (Best Sellers)
print("--- Top 5 Sản phẩm bán chạy nhất (theo số lượng) ---")
order_items.groupBy("product_id") \
  .agg(count("order_id").alias("quantity_sold")) \
  .orderBy(desc("quantity_sold")) \
  .show(5, truncate=False)


from pyspark.sql.functions import col, count, when, isnan

# =========================================================
# PHẦN 2: LÀM SẠCH DỮ LIỆU BẢNG ORDER ITEMS (ĐÃ SỬA LỖI)
# =========================================================
print("\n================ LÀM SẠCH BẢNG ORDER ITEMS ================")

# 1. Kiểm tra và Xử lý Trùng lặp
print("--- 1. Xử lý Dữ liệu Trùng lặp ---")
order_items_no_dups = order_items.dropDuplicates(["order_id", "order_item_id"])

new_items_count = order_items_no_dups.count()
print(f"Số dòng ban đầu: {total_items_sold}")
print(f"Số dòng sau khi xóa trùng: {new_items_count}")
print(f"Đã phát hiện và xóa: {total_items_sold - new_items_count} dòng trùng lặp.\n")

# 2. Kiểm tra và Xử lý Null (ĐÃ FIX LỖI)
print("--- 2. Kiểm tra Dữ liệu Null ---")
# Lấy kiểu dữ liệu của từng cột để phân loại
dtypes_dict = dict(order_items_no_dups.dtypes)

null_exprs = []
for c in order_items_no_dups.columns:
    if dtypes_dict[c] in ['double', 'float']:
        # Nếu là kiểu số thực, dùng cả isNull và isnan
        null_exprs.append(count(when(col(c).isNull() | isnan(c), c)).alias(c))
    else:
        # Nếu là chuỗi, thời gian (timestamp), số nguyên... thì chỉ dùng isNull
        null_exprs.append(count(when(col(c).isNull(), c)).alias(c))

# Hiển thị kết quả đếm Null
order_items_no_dups.select(null_exprs).show()

print("--- Xử lý Dữ liệu Null ---")
# Cắt bỏ các dòng bị thiếu thông tin quan trọng
order_items_cleaned = order_items_no_dups.dropna(subset=["order_id", "product_id", "price"])

final_items_count = order_items_cleaned.count()
print(f"--- Hoàn tất! Số dòng dữ liệu order_items_cleaned cuối cùng: {final_items_count} ---")
```

```python
from pyspark.sql.functions import col, count, desc, sum, avg, max, when, isnan

# =========================================================
# PHẦN 1: KHÁM PHÁ DỮ LIỆU BẢNG ORDER PAYMENTS
# =========================================================
print("================ KHÁM PHÁ BẢNG ORDER PAYMENTS ================")
print("--- Cấu trúc dữ liệu ---")
order_payments.printSchema()
order_payments.show(5, truncate=False)

total_payment_records = order_payments.count()
unique_paid_orders = order_payments.select("order_id").distinct().count()

print(f"Tổng số lượt giao dịch thanh toán: {total_payment_records}")
print(f"Tổng số đơn hàng thực tế đã thanh toán: {unique_paid_orders}")
print(f"Số lượng giao dịch dùng nhiều phương thức thanh toán/voucher: {total_payment_records - unique_paid_orders}")
print("-" * 40)

# Xem Phương thức thanh toán nào được ưa chuộng nhất
print("--- Tỷ lệ các Phương thức thanh toán ---")
order_payments.groupBy("payment_type") \
  .agg(
      count("order_id").alias("total_transactions"),
      sum("payment_value").alias("total_value")
  ) \
  .orderBy(desc("total_transactions")) \
  .show(truncate=False)

# Phân tích Giá trị thanh toán và Trả góp
avg_value = order_payments.select(avg("payment_value")).collect()[0][0]
max_installments = order_payments.select(max("payment_installments")).collect()[0][0]

print(f"--- THỐNG KÊ NHANH ---")
print(f"Giá trị thanh toán trung bình mỗi giao dịch: {avg_value:.2f} BRL")
print(f"Số tháng trả góp tối đa khách hàng từng chọn: {max_installments} tháng")


# =========================================================
# PHẦN 2: LÀM SẠCH DỮ LIỆU BẢNG ORDER PAYMENTS
# =========================================================
print("\n================ LÀM SẠCH BẢNG ORDER PAYMENTS ================")

# 1. Kiểm tra và Xử lý Trùng lặp
print("--- 1. Xử lý Dữ liệu Trùng lặp ---")
# Khóa chính của bảng này là kết hợp giữa [Mã đơn hàng] và [Thứ tự thanh toán]
payments_no_dups = order_payments.dropDuplicates(["order_id", "payment_sequential"])

new_payment_count = payments_no_dups.count()
print(f"Số dòng ban đầu: {total_payment_records}")
print(f"Số dòng sau khi xóa trùng: {new_payment_count}")
print(f"Đã phát hiện và xóa: {total_payment_records - new_payment_count} dòng trùng lặp.\n")

# 2. Kiểm tra và Xử lý Null (Dùng cách an toàn đã fix lỗi từ trước)
print("--- 2. Kiểm tra Dữ liệu Null ---")
dtypes_dict = dict(payments_no_dups.dtypes)

null_exprs = []
for c in payments_no_dups.columns:
    if dtypes_dict[c] in ['double', 'float']:
        null_exprs.append(count(when(col(c).isNull() | isnan(c), c)).alias(c))
    else:
        null_exprs.append(count(when(col(c).isNull(), c)).alias(c))

payments_no_dups.select(null_exprs).show()

print("--- Xử lý Dữ liệu Null ---")
# Vì đây là dữ liệu tài chính, nếu thiếu giá trị thanh toán hoặc mã đơn thì dòng đó trở nên vô nghĩa.
# Nếu phát hiện Null ở các cột quan trọng, ta sẽ cắt bỏ dòng đó.
order_payments_cleaned = payments_no_dups.dropna(subset=["order_id", "payment_value", "payment_type"])

# Đôi khi có những dòng payment_value = 0 (có thể do lỗi hệ thống), ta có thể lọc bỏ luôn
order_payments_cleaned = order_payments_cleaned.filter(col("payment_value") > 0)

final_payment_count = order_payments_cleaned.count()
print(f"--- Hoàn tất! Số dòng dữ liệu order_payments_cleaned cuối cùng: {final_payment_count} ---")
```

```python
# CHÍNH DÒNG NÀY ĐỊNH NGHĨA spark_round
from pyspark.sql.functions import col, datediff, when, round as spark_round, max as spark_max, countDistinct, sum as spark_sum, lit

print("1. Đang tạo các Derived Columns cho Classification & Regression...")

# SỬA ĐÚNG 1 CHỖ: Đổi df_final thành df_final_cleaned
df_features = (df_final_cleaned
    # 1. Thời gian giao hàng thực tế (tính bằng ngày)
    .withColumn("delivery_time_days",
                datediff(col("order_delivered_customer_date"), col("order_purchase_timestamp")))

    # 2. Khách hàng có bị giao trễ so với dự kiến không? (1 = Trễ, 0 = Không trễ/Chưa giao)
    .withColumn("is_late_delivery",
                when(col("order_delivered_customer_date") > col("order_estimated_delivery_date"), 1)
                .otherwise(0))

    # 3. Thể tích sản phẩm (cm3)
    .withColumn("product_volume_cm3",
                col("product_length_cm") * col("product_height_cm") * col("product_width_cm"))

    # 4. Tổng giá trị mỗi item
    .withColumn("total_item_value", spark_round(col("price") + col("freight_value"), 2))
)

# Hiển thị kết quả kiểm tra
df_features.select("order_id", "delivery_time_days", "is_late_delivery", "product_volume_cm3", "total_item_value").show(5)
```

```python
from pyspark.ml import Pipeline
from pyspark.ml.feature import StringIndexer, OneHotEncoder, VectorAssembler, StandardScaler

print("================ XÂY DỰNG REGRESSION PIPELINE (ĐỘC LẬP & CHUẨN KHOA HỌC) ================")

# 1. Xác định Label (Biến mục tiêu)
reg_label = "freight_value"

# 2. Lựa chọn đặc trưng (Feature Selection)
# ---> CẬP NHẬT: Đã đưa Thể tích (product_volume_cm3) và Thời gian giao hàng (delivery_time_days) vào
numeric_cols_reg = [
    "price",
    "product_weight_g",
    "product_volume_cm3",
    "delivery_time_days"
]

categorical_cols_reg = ["payment_type", "customer_state", "order_status"]

stages_reg = []

# --- BƯỚC 3: XỬ LÝ BIẾN PHÂN LOẠI (CATEGORICAL) ---
encoded_cat_cols = []
for cat_col in categorical_cols_reg:
    # Fill Null an toàn để tránh lỗi
    df_features = df_features.fillna({cat_col: "Unknown"})

    indexer = StringIndexer(inputCol=cat_col, outputCol=cat_col + "_idx", handleInvalid="keep")
    encoder = OneHotEncoder(inputCol=cat_col + "_idx", outputCol=cat_col + "_vec")
    stages_reg += [indexer, encoder]
    encoded_cat_cols.append(cat_col + "_vec")

# --- BƯỚC 4: TỔNG HỢP TOÀN BỘ ĐẶC TRƯNG ---
final_assembler_reg = VectorAssembler(
    inputCols=numeric_cols_reg + encoded_cat_cols,
    outputCol="unscaled_features_reg",
    handleInvalid="skip"
)
stages_reg += [final_assembler_reg]

# --- BƯỚC 5: CHUẨN HÓA DỮ LIỆU (StandardScaler) ---
scaler_reg = StandardScaler(
    inputCol="unscaled_features_reg",
    outputCol="final_features_reg",
    withStd=True,
    withMean=False
)
stages_reg += [scaler_reg]

# --- BƯỚC 6: KHỞI TẠO VÀ HUẤN LUYỆN PIPELINE HỒI QUY ---
print("Đang fit dữ liệu vào Regression Pipeline...")
pipeline_reg = Pipeline(stages=stages_reg)

# Fit và Transform dữ liệu
pipeline_reg_model = pipeline_reg.fit(df_features)
df_reg_output = pipeline_reg_model.transform(df_features)

print("✅ Pipeline Hồi quy ĐỘC LẬP đã sẵn sàng!")
df_reg_output.select(reg_label, "final_features_reg").show(5, truncate=False)
```

```python
import time
from pyspark.ml.regression import LinearRegression, DecisionTreeRegressor, RandomForestRegressor
from pyspark.ml.evaluation import RegressionEvaluator

print("================ HUẤN LUYỆN VÀ ĐÁNH GIÁ 3 MÔ HÌNH REGRESSION ================")

# 1. Đảm bảo dữ liệu không có Null ở cột Target (freight_value)
# Bước này cực kỳ quan trọng để các hàm tính toán Loss không bị lỗi
df_reg_model = df_reg_output.dropna(subset=[reg_label])

# 2. Train/Test Split (80/20)
train_reg, test_reg = df_reg_model.randomSplit([0.8, 0.2], seed=42)
print(f"Số dòng tập Train: {train_reg.count()} | Số dòng tập Test: {test_reg.count()}")

# 3. Khởi tạo 3 Models với Hyperparameters cụ thể
# Linear Regression: Giữ nguyên vì kết quả đã khá tốt
lr = LinearRegression(featuresCol="final_features_reg", labelCol=reg_label, maxIter=100, regParam=0.1, elasticNetParam=0.0)

# Decision Tree: Tăng maxDepth từ 5 lên 10 để cây học được nhiều quy luật hơn
dt = DecisionTreeRegressor(featuresCol="final_features_reg", labelCol=reg_label, maxDepth=10, maxBins=32)

# Random Forest: Tăng numTrees (số lượng cây) lên 100 và maxDepth lên 10
# Đây là mô hình kỳ vọng sẽ cho kết quả tốt nhất sau khi chỉnh
rf = RandomForestRegressor(featuresCol="final_features_reg", labelCol=reg_label, numTrees=100, maxDepth=10)
models_reg = {
    "Linear Regression": lr,
    "Decision Tree Regressor": dt,
    "Random Forest Regressor": rf
}

# 4. Cấu hình Evaluator cho Regression (RMSE, MAE, R2)
evaluator_rmse = RegressionEvaluator(labelCol=reg_label, predictionCol="prediction", metricName="rmse")
evaluator_mae = RegressionEvaluator(labelCol=reg_label, predictionCol="prediction", metricName="mae")
evaluator_r2 = RegressionEvaluator(labelCol=reg_label, predictionCol="prediction", metricName="r2")

results_reg = []

# 5. Vòng lặp Huấn luyện và Đánh giá
for name, model in models_reg.items():
    print(f"\n[+] Đang huấn luyện {name}...")
    start_time = time.time()

    # Huấn luyện mô hình
    fitted_model = model.fit(train_reg)
    train_time = time.time() - start_time

    # Dự đoán trên tập Test
    predictions = fitted_model.transform(test_reg)

    # Tính toán các Metrics
    rmse = evaluator_rmse.evaluate(predictions)
    mae = evaluator_mae.evaluate(predictions)
    r2 = evaluator_r2.evaluate(predictions)

    # Lưu kết quả
    results_reg.append((name, round(rmse, 4), round(mae, 4), round(r2, 4), round(train_time, 2)))
    print(f"    -> Hoàn tất trong {train_time:.2f} giây! RMSE: {rmse:.4f} | R2: {r2:.4f}")

# 6. In Bảng Kết Quả Cuối Cùng
print("\n================ BẢNG TỔNG HỢP KẾT QUẢ ================")
results_reg_df = spark.createDataFrame(results_reg, ["Model", "RMSE", "MAE", "R2", "Train_time_sec"])
results_reg_df.show(truncate=False)
```
